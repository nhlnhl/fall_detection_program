```python
import pandas as pd
import numpy as np
import math
import warnings
warnings.filterwarnings('ignore')

raw_data = pd.read_csv("CompleteDataSet_training_competition_1500.csv", sep=",")

data = raw_data[['AnkleAccelerometer x-axis', 'AnkleAccelerometer y-axis', 'AnkleAccelerometer z-axis', 'BeltAccelerometer x-axis', 'BeltAccelerometer y-axis', 'BeltAccelerometer z-axis', 'WristAccelerometer x-axis', 'WristAccelerometer y-axis', 'WristAccelerometer z-axis', 'BrainSensor', 'Activity']]
av = raw_data[['AnkleAngularVelocity x-axis', 'AnkleAngularVelocity y-axis', 'AnkleAngularVelocity z-axis', 'BeltAngularVelocity x-axis', 'BeltAngularVelocity y-axis', 'BeltAngularVelocity z-axis', 'WristAngularVelocity x-axis', 'WristAngularVelocity y-axis', 'WristAngularVelocity z-axis']]

n_step = 1500
n_batch = int(data.shape[0] / n_step)

# variables used in the paper
k = 15
t = 3
K = 300

xyz = 3

data = data.as_matrix()
av = av.as_matrix()
calculate = np.zeros([n_batch * int(k / t), 138 + 1])

# for each data
for i in range(n_batch):
    
    # for each sliding window
    for j in range(int(k / t)):
        calculate[int(k / t) * i + j][138] = data[i * n_step][xyz+xyz+xyz+1]
        
        # ankle variables
        ankle_x = np.zeros([K])
        ankle_y = np.zeros([K])
        ankle_z = np.zeros([K])
        ankle_a = np.zeros([K])
        ankle_sa = 0
        ankle_se = np.zeros([xyz])
        fft_ankle_x = np.zeros([K])
        fft_ankle_y = np.zeros([K])
        fft_ankle_z = np.zeros([K])
        fft_ankle_a = np.zeros([K])
        sqr_ankle_x = np.zeros([K])
        sqr_ankle_y = np.zeros([K])
        sqr_ankle_z = np.zeros([K])
        sqr_ankle_a = np.zeros([K])
        ankle_av = np.zeros([xyz+1])
        psd_ankle_x = 0
        psd_ankle_y = 0
        psd_ankle_z = 0
        psd_ankle_a = 0
        ankle_entropy = np.zeros([xyz+1])
        ankle_corr = np.zeros((xyz, xyz))
        
        # belt variables
        belt_x = np.zeros([K])
        belt_y = np.zeros([K])
        belt_z = np.zeros([K])
        belt_a = np.zeros([K])
        belt_sa = 0
        belt_se = np.zeros([xyz])
        fft_belt_x = np.zeros([K])
        fft_belt_y = np.zeros([K])
        fft_belt_z = np.zeros([K])
        fft_belt_a = np.zeros([K])
        sqr_belt_x = np.zeros([K])
        sqr_belt_y = np.zeros([K])
        sqr_belt_z = np.zeros([K])
        sqr_belt_a = np.zeros([K])
        belt_av = np.zeros([xyz+1])
        psd_belt_x = 0
        psd_belt_y = 0
        psd_belt_z = 0
        psd_belt_a = 0
        belt_entropy = np.zeros([xyz+1])
        belt_corr = np.zeros((xyz, xyz))
        
        # wrist variables
        wrist_x = np.zeros([K])
        wrist_y = np.zeros([K])
        wrist_z = np.zeros([K])
        wrist_a = np.zeros([K])
        wrist_sa = 0
        wrist_se = np.zeros([xyz])
        fft_wrist_x = np.zeros([K])
        fft_wrist_y = np.zeros([K])
        fft_wrist_z = np.zeros([K])
        fft_wrist_a = np.zeros([K])
        sqr_wrist_x = np.zeros([K])
        sqr_wrist_y = np.zeros([K])
        sqr_wrist_z = np.zeros([K])
        sqr_wrist_a = np.zeros([K])
        wrist_av = np.zeros([xyz+1])
        psd_wrist_x = 0
        psd_wrist_y = 0
        psd_wrist_z = 0
        psd_wrist_a = 0
        wrist_entropy = np.zeros([xyz+1])
        wrist_corr = np.zeros((xyz, xyz))
        
        # brainwave variables
        brainwave = np.zeros([K])
        brainwave_sa = 0
        fft_brainwave = np.zeros([K])
        sqr_brainwave = np.zeros([K])
        psd_brainwave = 0
        brainwave_entropy = 0     
        
        for m in range(K):                
            ankle_x[m] = data[i * n_step + j * K + m][0]
            ankle_y[m] = data[i * n_step + j * K + m][1]
            ankle_z[m] = data[i * n_step + j * K + m][2]
            ankle_a[m] = np.sqrt(pow(data[i * n_step + j * K + m][0], 2) + pow(data[i * n_step + j * K + m][1], 2) + pow(data[i * n_step + j * K + m][2], 2))
            ankle_sa += (np.abs(ankle_x[m]) + np.abs(ankle_y[m]) + np.abs(ankle_z[m]))
            ankle_av[0] += av[i * n_step + j * K + m][0]
            ankle_av[1] += av[i * n_step + j * K + m][1]
            ankle_av[2] += av[i * n_step + j * K + m][2]
            ankle_av[3] += np.sqrt(pow(av[i * n_step + j * K + m][0], 2) + pow(av[i * n_step + j * K + m][1], 2) + pow(av[i * n_step + j * K + m][2], 2))
            belt_x[m] = data[i * n_step + j * K + m][3]
            belt_y[m] = data[i * n_step + j * K + m][4]
            belt_z[m] = data[i * n_step + j * K + m][5]
            belt_a[m] = np.sqrt(pow(data[i * n_step + j * K + m][3], 2) + pow(data[i * n_step + j * K + m][4], 2) + pow(data[i * n_step + j * K + m][5], 2))
            belt_sa += (np.abs(belt_x[m]) + np.abs(belt_y[m]) + np.abs(belt_z[m]))
            belt_av[0] += av[i * n_step + j * K + m][3]
            belt_av[1] += av[i * n_step + j * K + m][4]
            belt_av[2] += av[i * n_step + j * K + m][5]
            belt_av[3] += np.sqrt(pow(av[i * n_step + j * K + m][3], 2) + pow(av[i * n_step + j * K + m][4], 2) + pow(av[i * n_step + j * K + m][5], 2))
            wrist_x[m] = data[i * n_step + j * K + m][6]
            wrist_y[m] = data[i * n_step + j * K + m][7]
            wrist_z[m] = data[i * n_step + j * K + m][8]
            wrist_a[m] = np.sqrt(pow(data[i * n_step + j * K + m][6], 2) + pow(data[i * n_step + j * K + m][7], 2) + pow(data[i * n_step + j * K + m][8], 2))
            wrist_sa += (np.abs(wrist_x[m]) + np.abs(wrist_y[m]) + np.abs(wrist_z[m]))
            wrist_av[0] += av[i * n_step + j * K + m][6]
            wrist_av[1] += av[i * n_step + j * K + m][7]
            wrist_av[2] += av[i * n_step + j * K + m][8]
            wrist_av[3] += np.sqrt(pow(av[i * n_step + j * K + m][6], 2) + pow(av[i * n_step + j * K + m][7], 2) + pow(av[i * n_step + j * K + m][8], 2))
            brainwave[m] = data[i * n_step + j * K + m][9]
            brainwave_sa += np.abs(brainwave[m])
        
        fft_ankle_x = np.fft.fft(ankle_x)
        fft_ankle_y = np.fft.fft(ankle_y)
        fft_ankle_z = np.fft.fft(ankle_z)
        fft_ankle_a = np.fft.fft(ankle_a)
        fft_belt_x = np.fft.fft(belt_x)
        fft_belt_y = np.fft.fft(belt_y)
        fft_belt_z = np.fft.fft(belt_z)
        fft_belt_a = np.fft.fft(belt_a)
        fft_wrist_x = np.fft.fft(wrist_x)
        fft_wrist_y = np.fft.fft(wrist_y)
        fft_wrist_z = np.fft.fft(wrist_z)
        fft_wrist_a = np.fft.fft(wrist_a)
        fft_brainwave = np.fft.fft(brainwave)
        for m in range(K):
            ankle_se[0] += pow(fft_ankle_x[m], 2)
            ankle_se[1] += pow(fft_ankle_y[m], 2)
            ankle_se[2] += pow(fft_ankle_z[m], 2)
            belt_se[0] += pow(fft_belt_x[m], 2)
            belt_se[1] += pow(fft_belt_y[m], 2)
            belt_se[2] += pow(fft_belt_z[m], 2)
            wrist_se[0] += pow(fft_wrist_x[m], 2)
            wrist_se[1] += pow(fft_wrist_y[m], 2)
            wrist_se[2] += pow(fft_wrist_z[m], 2)
            
            sqr_ankle_x[m] = np.sqrt(fft_ankle_x[m])
            sqr_ankle_y[m] = np.sqrt(fft_ankle_y[m])
            sqr_ankle_z[m] = np.sqrt(fft_ankle_z[m])
            sqr_ankle_a[m] = np.sqrt(fft_ankle_a[m])
            sqr_belt_x[m] = np.sqrt(fft_belt_x[m])
            sqr_belt_y[m] = np.sqrt(fft_belt_y[m])
            sqr_belt_z[m] = np.sqrt(fft_belt_z[m])
            sqr_belt_a[m] = np.sqrt(fft_belt_a[m])
            sqr_wrist_x[m] = np.sqrt(fft_wrist_x[m])
            sqr_wrist_y[m] = np.sqrt(fft_wrist_y[m])
            sqr_wrist_z[m] = np.sqrt(fft_wrist_z[m])
            sqr_wrist_a[m] = np.sqrt(fft_wrist_a[m])
            sqr_brainwave = np.sqrt(fft_brainwave)
        
        psd_ankle_x = np.mean(sqr_ankle_x)
        psd_ankle_y = np.mean(sqr_ankle_y)
        psd_ankle_z = np.mean(sqr_ankle_z)
        psd_ankle_a = np.mean(sqr_ankle_a)
        psd_belt_x = np.mean(sqr_belt_x)
        psd_belt_y = np.mean(sqr_belt_y)
        psd_belt_z = np.mean(sqr_belt_z)
        psd_belt_a = np.mean(sqr_belt_a)
        psd_wrist_x = np.mean(sqr_wrist_x)
        psd_wrist_y = np.mean(sqr_wrist_y)
        psd_wrist_z = np.mean(sqr_wrist_z)
        psd_wrist_a = np.mean(sqr_wrist_a)
        psd_brainwave = np.mean(sqr_brainwave)
        ankle_entropy[0] = psd_ankle_x * np.log2(psd_ankle_x)
        ankle_entropy[1] = psd_ankle_y * np.log2(psd_ankle_y)
        ankle_entropy[2] = psd_ankle_z * np.log2(psd_ankle_z)
        ankle_entropy[3] = psd_ankle_a * np.log2(psd_ankle_a)
        belt_entropy[0] = psd_belt_x * np.log2(psd_belt_x)
        belt_entropy[1] = psd_belt_y * np.log2(psd_belt_y)
        belt_entropy[2] = psd_belt_z * np.log2(psd_belt_z)
        belt_entropy[3] = psd_belt_a * np.log2(psd_belt_a)
        wrist_entropy[0] = psd_wrist_x * np.log2(psd_wrist_x)
        wrist_entropy[1] = psd_wrist_y * np.log2(psd_wrist_y)
        wrist_entropy[2] = psd_wrist_z * np.log2(psd_wrist_z)
        wrist_entropy[3] = psd_wrist_a * np.log2(psd_wrist_a)
        brainwave_entropy = psd_brainwave * np.log2(psd_brainwave)
        
        ankle_lst = [ankle_x, ankle_y, ankle_z]
        ankle_df = pd.DataFrame(ankle_lst).T
        ankle_corr = ankle_df.corr(method = 'pearson')
        belt_lst = [belt_x, belt_y, belt_z]
        belt_df = pd.DataFrame(belt_lst).T
        belt_corr = belt_df.corr(method = 'pearson')
        wrist_lst = [wrist_x, wrist_y, wrist_z]
        wrist_df = pd.DataFrame(wrist_lst).T
        wrist_corr = wrist_df.corr(method = 'pearson')
        
        calculate[int(k / t) * i + j][0] = np.std(ankle_x)
        calculate[int(k / t) * i + j][1] = np.std(ankle_y)
        calculate[int(k / t) * i + j][2] = np.std(ankle_z)
        calculate[int(k / t) * i + j][3] = np.std(ankle_a)
        calculate[int(k / t) * i + j][4] = np.mean(ankle_x)
        calculate[int(k / t) * i + j][5] = np.mean(ankle_y)
        calculate[int(k / t) * i + j][6] = np.mean(ankle_z)
        calculate[int(k / t) * i + j][7] = np.mean(ankle_a)
        calculate[int(k / t) * i + j][8] = max(ankle_x)
        calculate[int(k / t) * i + j][9] = max(ankle_y)
        calculate[int(k / t) * i + j][10] = max(ankle_z)
        calculate[int(k / t) * i + j][11] = max(ankle_a)
        calculate[int(k / t) * i + j][12] = min(ankle_x)
        calculate[int(k / t) * i + j][13] = min(ankle_y)
        calculate[int(k / t) * i + j][14] = min(ankle_z)
        calculate[int(k / t) * i + j][15] = min(ankle_a)
        calculate[int(k / t) * i + j][16] = ankle_sa
        calculate[int(k / t) * i + j][17] = np.median(ankle_x)
        calculate[int(k / t) * i + j][18] = np.median(ankle_y)
        calculate[int(k / t) * i + j][19] = np.median(ankle_z)
        calculate[int(k / t) * i + j][20] = np.median(ankle_a)
        calculate[int(k / t) * i + j][21] = np.percentile(ankle_x, 25)
        calculate[int(k / t) * i + j][22] = np.percentile(ankle_y, 25)
        calculate[int(k / t) * i + j][23] = np.percentile(ankle_z, 25)
        calculate[int(k / t) * i + j][24] = np.percentile(ankle_a, 25)
        calculate[int(k / t) * i + j][25] = np.percentile(ankle_x, 75)
        calculate[int(k / t) * i + j][26] = np.percentile(ankle_y, 75)
        calculate[int(k / t) * i + j][27] = np.percentile(ankle_z, 75)
        calculate[int(k / t) * i + j][28] = np.percentile(ankle_a, 75)
        calculate[int(k / t) * i + j][29] = ankle_se[0]
        calculate[int(k / t) * i + j][30] = ankle_se[1]
        calculate[int(k / t) * i + j][31] = ankle_se[2]
        calculate[int(k / t) * i + j][32] = ankle_av[0] / K
        calculate[int(k / t) * i + j][33] = ankle_av[1] / K
        calculate[int(k / t) * i + j][34] = ankle_av[2] / K
        calculate[int(k / t) * i + j][35] = ankle_av[3] / K
        calculate[int(k / t) * i + j][36] = ankle_entropy[0]
        calculate[int(k / t) * i + j][37] = ankle_entropy[1]
        calculate[int(k / t) * i + j][38] = ankle_entropy[2]
        calculate[int(k / t) * i + j][39] = ankle_entropy[3]
        calculate[int(k / t) * i + j][40] = ankle_corr[0][1]
        calculate[int(k / t) * i + j][41] = ankle_corr[1][2]
        calculate[int(k / t) * i + j][42] = ankle_corr[0][2]        
        calculate[int(k / t) * i + j][43] = np.std(belt_x)
        calculate[int(k / t) * i + j][44] = np.std(belt_y)
        calculate[int(k / t) * i + j][45] = np.std(belt_z)
        calculate[int(k / t) * i + j][46] = np.std(belt_a)
        calculate[int(k / t) * i + j][47] = np.mean(belt_x)
        calculate[int(k / t) * i + j][48] = np.mean(belt_y)
        calculate[int(k / t) * i + j][49] = np.mean(belt_z)
        calculate[int(k / t) * i + j][50] = np.mean(belt_a)
        calculate[int(k / t) * i + j][51] = max(belt_x)
        calculate[int(k / t) * i + j][52] = max(belt_y)
        calculate[int(k / t) * i + j][53] = max(belt_z)
        calculate[int(k / t) * i + j][54] = max(belt_a)
        calculate[int(k / t) * i + j][55] = min(belt_x)
        calculate[int(k / t) * i + j][56] = min(belt_y)
        calculate[int(k / t) * i + j][57] = min(belt_z)
        calculate[int(k / t) * i + j][58] = min(belt_a)
        calculate[int(k / t) * i + j][59] = belt_sa
        calculate[int(k / t) * i + j][60] = np.median(belt_x)
        calculate[int(k / t) * i + j][61] = np.median(belt_y)
        calculate[int(k / t) * i + j][62] = np.median(belt_z)
        calculate[int(k / t) * i + j][63] = np.median(belt_a)
        calculate[int(k / t) * i + j][64] = np.percentile(belt_x, 25)
        calculate[int(k / t) * i + j][65] = np.percentile(belt_y, 25)
        calculate[int(k / t) * i + j][66] = np.percentile(belt_z, 25)
        calculate[int(k / t) * i + j][67] = np.percentile(belt_a, 25)
        calculate[int(k / t) * i + j][68] = np.percentile(belt_x, 75)
        calculate[int(k / t) * i + j][69] = np.percentile(belt_y, 75)
        calculate[int(k / t) * i + j][70] = np.percentile(belt_z, 75)
        calculate[int(k / t) * i + j][71] = np.percentile(belt_a, 75)
        calculate[int(k / t) * i + j][72] = belt_se[0]
        calculate[int(k / t) * i + j][73] = belt_se[1]
        calculate[int(k / t) * i + j][74] = belt_se[2]
        calculate[int(k / t) * i + j][75] = belt_av[0] / K
        calculate[int(k / t) * i + j][76] = belt_av[1] / K
        calculate[int(k / t) * i + j][77] = belt_av[2] / K
        calculate[int(k / t) * i + j][78] = belt_av[3] / K
        calculate[int(k / t) * i + j][79] = belt_entropy[0]
        calculate[int(k / t) * i + j][80] = belt_entropy[1]
        calculate[int(k / t) * i + j][81] = belt_entropy[2]
        calculate[int(k / t) * i + j][82] = belt_entropy[3]
        calculate[int(k / t) * i + j][83] = belt_corr[0][1]
        calculate[int(k / t) * i + j][84] = belt_corr[1][2]
        calculate[int(k / t) * i + j][85] = belt_corr[0][2]
        calculate[int(k / t) * i + j][86] = np.std(wrist_x)
        calculate[int(k / t) * i + j][87] = np.std(wrist_y)
        calculate[int(k / t) * i + j][88] = np.std(wrist_z)
        calculate[int(k / t) * i + j][89] = np.std(wrist_a)
        calculate[int(k / t) * i + j][90] = np.mean(wrist_x)
        calculate[int(k / t) * i + j][91] = np.mean(wrist_y)
        calculate[int(k / t) * i + j][92] = np.mean(wrist_z)
        calculate[int(k / t) * i + j][93] = np.mean(wrist_a)
        calculate[int(k / t) * i + j][94] = max(wrist_x)
        calculate[int(k / t) * i + j][95] = max(wrist_y)
        calculate[int(k / t) * i + j][96] = max(wrist_z)
        calculate[int(k / t) * i + j][97] = max(wrist_a)
        calculate[int(k / t) * i + j][98] = min(wrist_x)
        calculate[int(k / t) * i + j][99] = min(wrist_y)
        calculate[int(k / t) * i + j][100] = min(wrist_z)
        calculate[int(k / t) * i + j][101] = min(wrist_a)
        calculate[int(k / t) * i + j][102] = wrist_sa
        calculate[int(k / t) * i + j][103] = np.median(wrist_x)
        calculate[int(k / t) * i + j][104] = np.median(wrist_y)
        calculate[int(k / t) * i + j][105] = np.median(wrist_z)
        calculate[int(k / t) * i + j][106] = np.median(wrist_a)
        calculate[int(k / t) * i + j][107] = np.percentile(wrist_x, 25)
        calculate[int(k / t) * i + j][108] = np.percentile(wrist_y, 25)
        calculate[int(k / t) * i + j][109] = np.percentile(wrist_z, 25)
        calculate[int(k / t) * i + j][110] = np.percentile(wrist_a, 25)
        calculate[int(k / t) * i + j][111] = np.percentile(wrist_x, 75)
        calculate[int(k / t) * i + j][112] = np.percentile(wrist_y, 75)
        calculate[int(k / t) * i + j][113] = np.percentile(wrist_z, 75)
        calculate[int(k / t) * i + j][114] = np.percentile(wrist_a, 75)
        calculate[int(k / t) * i + j][115] = wrist_se[0]
        calculate[int(k / t) * i + j][116] = wrist_se[1]
        calculate[int(k / t) * i + j][117] = wrist_se[2]
        calculate[int(k / t) * i + j][118] = wrist_av[0] / K
        calculate[int(k / t) * i + j][119] = wrist_av[1] / K
        calculate[int(k / t) * i + j][120] = wrist_av[2] / K
        calculate[int(k / t) * i + j][121] = wrist_av[3] / K
        calculate[int(k / t) * i + j][122] = wrist_entropy[0]
        calculate[int(k / t) * i + j][123] = wrist_entropy[1]
        calculate[int(k / t) * i + j][124] = wrist_entropy[2]
        calculate[int(k / t) * i + j][125] = wrist_entropy[3]
        calculate[int(k / t) * i + j][126] = wrist_corr[0][1]
        calculate[int(k / t) * i + j][127] = wrist_corr[1][2]
        calculate[int(k / t) * i + j][128] = wrist_corr[0][2]
        calculate[int(k / t) * i + j][129] = np.std(brainwave)
        calculate[int(k / t) * i + j][130] = np.mean(brainwave)
        calculate[int(k / t) * i + j][131] = max(brainwave)
        calculate[int(k / t) * i + j][132] = min(brainwave)
        calculate[int(k / t) * i + j][133] = brainwave_sa
        calculate[int(k / t) * i + j][134] = np.median(brainwave)
        calculate[int(k / t) * i + j][135] = np.percentile(brainwave, 25)
        calculate[int(k / t) * i + j][136] = np.percentile(brainwave, 75)
        calculate[int(k / t) * i + j][137] = brainwave_entropy
        
data = pd.DataFrame(calculate)

data.to_csv("./dataset/training_data_LSTM_FS_3_AnkleBeltWristBW.csv", sep=",", header=False, index=False)
```
